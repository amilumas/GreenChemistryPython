```python
def power2(num, raisedto = 2): 
    '''
    This function takes the value specified in the 'num' parameter and raises it to 
    the power of the number specified in the 'rasiedto' parameter and returns the result
    '''
    # Only works if raisedto is an integer
    x = 1
    for i in range(raisedto):
        x = x * num
    return x

    
```

```python
def plotselectedsolvents(solventdf, xlabel, ylabel, bboxtoanchor=(0.5,-1)):
    '''Plots solvents individually with x and y attributes
    solventdf is a dataframe containing solvent data, xlabel and ylabel are columns
    '''
    colormap = plt.cm.nipy_spectral
    number_of_plots = solventdf.shape[0]
    fig = plt.figure()
    ax1 = fig.add_subplot(111)
    ax1.set_color_cycle([colormap(i) for i in np.linspace(0, 1, number_of_plots)])
    for index, row in solventdf.iterrows():
        ax1.scatter(row[xlabel], row[ylabel], label = index)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.legend(loc='lower center', ncol=2, bbox_to_anchor=bboxtoanchor)
    plt.show()
```

```python
plotselectedsolvents(selsolvents, 'MP', 'BP')
```
